### Compare June and December data across all years 

```python
from scipy import stats
```

```python
# Filter data for desired months
df = pd.read_csv('Resources/hawaii_measurements.csv')
df.date=pd.to_datetime(df.date)
df1=df.loc[df.date.dt.month==6]
display(df1)
```

```python
# Identify the average temperature for June
june_mean=df1.tobs.mean()
print(june_mean)
```

```python
#get statistic data for df1 to verify mean tobs
df1.describe()
```

```python
# Identify the average temperature for December
#filter data for December
df2=df.loc[df.date.dt.month==12]
display(df2)
#identify the average temperature for December
december_mean=df2.tobs.mean()
print(december_mean)
```

```python
#get statistic data for df1 to verify mean tobs
df2.describe()
```

```python
# Create collections of temperature data
june=df1.tobs
december=df2.tobs
print(june.var())
print(december.var())
```

```python
# Run independent t-test
stats.ttest_ind(june,december,equal_var=False)
```

### Analysis

```python
#There is significant difference between June temperature and December temperature.
#I use unpaired t-test because the two groups are independent. The number of two groups is different. One is 1700 and the other is 1517.
```
